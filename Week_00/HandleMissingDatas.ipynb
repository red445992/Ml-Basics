```python
import pandas as pd
```

```python
data = pd.read_csv("../data/weather_data.csv",parse_dates=["day"])
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

### Filling missing values with Zero

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({

    "temperature":0,
    "event":"no event"
})
new_df
```

```python
new_df = df.fillna(method = "ffill")
new_df
```

```python
new_df = df.interpolate()
new_df
```
