# categorial Variable

```python
names = ['grp1','grp2','grp3']
values = [10,50,100]

plt.subplot(131)
plt.bar(names,values)
plt.subplot(132)
plt.scatter(names,values)
plt.subplot(133)
plt.plot(names,values)
```
