# Data CLeaning

```python
# let us create a data set for now
```

```python
# 1. Empty cells
df = pd.read_csv('Housing.csv')
new_df = df.dropna()
new_df.head()
#dropna() method returns a new DataFrame and will not change the original
```

```python
df.dropna(inplace=True)
df.head()
```

```python
# Replace the empty values
# fillna() method allows us to replace empty cells with a value()
```

```python
df.fillna(130,inplace=True)
df.head()
```

```python
#Replace for only specified columns
df.fillna({"area":130},inplace=True)
df
```

```python
# Best way to replace empty cells is to calculate the mean, median or mode value of the column
```

```python
x = df['area'].mean()
df.fillna({'areas':x},inplace=True)
```

```python
y = df['area'].median()
df.fillna({'areas':y},inplace=True)
```

```python
z = df['bathrooms'].mode()[0]
df.fillna({'bathrooms':z},inplace=True)
df.head()
```

# cleaning Data of wrong format

```python
# we have to change the data strucutre of the data in a coulmn in a same format
# for that use to_string(), to_datetime()
```

```python
#if you want to remove the rows with null values
df.dropna(subset=['bedrooms'],inplace=True)
df
```

# fixing wrong data

```python
df.loc[7,'areas'] = 5
```

```python
df.head(10)
```

```python
# for large data sets its not possible for that we use lopp through all values in the duration
for x in df.index:
    if df.loc[x,"bedrooms"]>10:
        df.loc[x,'bedrooms'] = 9
df.head()
```

```python
# deleteing the rows where duration is higher than avg or max
for x in df.index:
    if  df.loc[x,"bedrooms"] > 10:
        df.drop(x,inplace = True)
df.head()
```
