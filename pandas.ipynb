# Data CLeaning

```python
# let us create a data set for now
```

```python
# 1. Empty cells
df = pd.read_csv('Housing.csv')
new_df = df.dropna()
new_df.head()
#dropna() method returns a new DataFrame and will not change the original
```

```python
df.dropna(inplace=True)
df.head()
```

```python
# Replace the empty values
# fillna() method allows us to replace empty cells with a value()
```

```python
df.fillna(130,inplace=True)
df.head()
```

```python
#Replace for only specified columns
df.fillna({"area":130},inplace=True)
df
```

```python
# Best way to replace empty cells is to calculate the mean, median or mode value of the column
```

```python
x = df['area'].mean()
df.fillna({'areas':x},inplace=True)
```

```python
y = df['area'].median()
df.fillna({'areas':y},inplace=True)
```

```python
z = df['bathrooms'].mode()[0]
df.fillna({'bathrooms':z},inplace=True)
df.head()
```
